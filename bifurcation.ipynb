```python
from pylab import *
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from matplotlib import rc

a = 0.4
b = 0.7

def func(x,t,a,b):
    dxdt = a - b*x
    return dxdt


xinit = [-2,-1,0,1,2]
t = np.linspace(0, 5, 100)
t1 = np.linspace(0, 3, 100)
X = np.linspace(-10, 10, 200)

y = a - b*X
y1 = 0*X

x = odeint(func, xinit, t, args=(a,b))

plt.figure(1)

plt.title('$Fixed \ points \ of \ \dot{x}=a - bx$', fontsize = 18)
plt.plot(X,y)
plt.plot(X,y1,'k-')
plt.plot(a/b,0,'ko')
plt.plot(0.45,0,'b>')
plt.plot(0.7,0,'b<')
plt.axis([0, 1, -0.5, 0.5])
plt.grid()
text(0.75, 0.4, "$a=%g$" %a, fontsize=17)
text(0.75, 0.3, "$b=%g$" %b, fontsize=17)
text(0.55, 0.05, "$x^*=a/b$", fontsize=17)
plt.xlabel('$x$',fontsize=18)
plt.ylabel('$f(x)$',fontsize=18)
plt.savefig('fixed1.pdf')
         
plt.figure(2)

plt.title('$Some \ trajectories \ of \ \dot{x}=a - bx$', fontsize = 18)
#for n in range(0,5):
plt.plot(t,x)
plt.plot(t,a/b + 0*t,'k-.')
plt.xlabel('$time$',fontsize=18)
plt.ylabel('$x(t)$',fontsize=18)
plt.grid()
plt.axis([0, 5, -2, 2])

plt.savefig('traj1.pdf')
plt.show()
```
