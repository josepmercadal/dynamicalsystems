# Pitchfork bifurcation

Consider the differential equation:

$$\dot{x}=\mu x - x^3$$

The point $x=0$ is a steady state, and since the jacobian determinant is $|J|= \mu - x^2$, it will be stable for $\mu < 0$ and unstable otherwise. The other fixed points are $x_{\pm}=\pm\sqrt{\mu}$, which only exist for $\mu > 0$ and are always stable. We can see this behaviour by plotting $f(x) = x$ and $f(x) = x^3/\mu$ and looking at the intersection points. We will consider three values of $\lambda =$ $-0.5$, $0$, and $0.5$. First we initialize and create the three figures:

```python
from pylab import *
import matplotlib.pyplot as plt
from scipy.integrate import odeint

fig, ((ax1, ax2, ax3)) = plt.subplots(1,3, sharey=True, figsize = (9,2.5))  # Three axes, horizontal, with shared y axis and the size of the figure
fig.subplots_adjust(wspace = 0.1)                                           # Space between plots 
```

In each <code>ax</code> we will plot a these functions for different $\mu$.

```python
X = np.linspace(-2, 2, 100)

for i in range(1,4):
    
    mu = 0.5*i - 1
    ax = globals()["ax" + str(i)]

    y = mu*X - X**3
    z = 0*X
    ax.plot(X,y,'k',label="$\mu=%g$" %mu)
    ax.plot(X,z,'k-')
    ax.legend(loc='best',fontsize = 12)
    ax.xaxis.set_ticks(np.arange(-2, 2.1, 1))
    ax.yaxis.set_ticks(np.arange(-0.3, 0.31, 0.1))
    
ax1.plot(0,0,'ko')
ax1.plot(-0.7,0,'k>')
ax1.plot(0.7,0,'k<')

ax2.plot(0,0,'ko')
ax2.plot(-0.7,0,'k>')
ax2.plot(0.7,0,'k<')

ax3.plot(-0.7,0,'ko', mfc = 'w')
ax3.plot(0.7,0,'ko', mfc='w')
ax3.plot(0,0,'ko')

ax3.plot(-1.1,0,'k<')
ax3.plot(-0.3,0,'k>')
ax3.plot(1.1,0,'k>')
ax3.plot(0.3,0,'k<')

plt.ylim([-0.3,0.3])

plt.show()

```

We will now plot the bifurcation diagram.

```python
plt.figure()

plt.ylabel('$x^*$',fontsize=20)
plt.xlabel('$\mu$',fontsize=20)
plt.axis([-0.5,1, -1, 1])

B = np.linspace(0,1,200)
C = np.linspace(-1,0,200)

F1 = sqrt(B)
F2 = -sqrt(B)

G = 0*B
H = 0*C

plt.plot(B,F1,'b', linewidth = 1.5)
plt.plot(B,F2,'b', linewidth = 1.5)
plt.plot(B,G,'b--', linewidth = 1.5)
plt.plot(C,H,'b-', linewidth = 1.5)

arrow(-0.2, 0.6, 0, -0.2, head_width=0.02, head_length=0.05, fc='k', ec='k')
arrow(-0.2, -0.6, 0, 0.2, head_width=0.02, head_length=0.05, fc='k', ec='k')
arrow(0.2, 0.8, 0, -0.2, head_width=0.02, head_length=0.05, fc='k', ec='k')
arrow(0.2, -0.8, 0, 0.2, head_width=0.02, head_length=0.05, fc='k', ec='k')
arrow(0.2, 0.1, 0, 0.2, head_width=0.02, head_length=0.05, fc='k', ec='k')
arrow(0.2, -0.1, 0, -0.2, head_width=0.02, head_length=0.05, fc='k', ec='k')
arrow(0.6, 0.3, 0, 0.2, head_width=0.02, head_length=0.05, fc='k', ec='k')
arrow(0.6, -0.3, 0, -0.2, head_width=0.02, head_length=0.05, fc='k', ec='k')

plt.show()
```
