```python
from numpy import *
import pylab as p

# Definition of parameters 

mu = 4

def dX_dt(X, t=0):
    """ Return the growth rate of fox and rabbit populations. """
    return array([ mu*(X[0] - (X[0]**3)/3 - X[1]) ,  
                  X[0]/mu ])

from scipy import integrate

t = linspace(0, 50,  1000)
X0 = array([1, 1])

X, infodict = integrate.odeint(dX_dt, X0, t, full_output=True)

x,y = X.T

f1 = p.figure()
p.plot(t, x, 'r-', label='X')
p.plot(t, y  , 'b-', label='Y')
p.grid()
p.legend(loc='best')
p.xlabel('time')
p.ylabel('$X,Y$')
p.ylim(-4, 4)
#f1.savefig('vanderpol1.png')

f2 = p.figure()
                 
#PHASE SPACE   

XX = linspace(-5,5,500)
YY = linspace(-5,5,500)

nc1 = XX - (XX**3)/3
nc2 = 0*XX
    
p.title('Trajectories in phase space')
p.plot(x,y)
p.plot(YY,nc1)
p.plot(YY,nc2)
p.xlabel('$X$')
p.ylabel('$Y$')
p.legend()
p.grid()
p.xlim(-3, 3)
p.ylim(-2, 2)
#f2.savefig('vanderpol2.png')
p.show()
```
